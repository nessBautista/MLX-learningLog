```python
# imports
from mlx_mlp import MLX_MNIST_MLP
from pytorch_mlp import TORCH_MNIST_MLP
from matplotlib import pyplot as plt
```

# Create a multi-layer perceptron with MLX and PyTorch

```python
# Create a multi-layer perceptron with MLX
mlx_mlp = MLX_MNIST_MLP(
    num_layers=2,
    hidden_dim=32,
    num_epochs=10,
    learning_rate=0.1
)
# Train the model
mlx_mlp.train()
# Get the metrics
mlx_metrics = mlx_mlp.get_metrics()
```

```python
# Create a multi-layer perceptron with PyTorch
torch_mlp = TORCH_MNIST_MLP(
    num_layers=2,
    hidden_dim=32,
    num_epochs=10,
    learning_rate=0.1
)
# Train the model
torch_mlp.train()
# Get the metrics
torch_metrics = torch_mlp.get_metrics()
```

# Plot the metrics and compare the the performance of MLX and PyTorch

```python

# Plot metrics
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# Plot training loss
ax1.plot(mlx_metrics['train_loss'], label='MLX')
ax1.plot(torch_metrics['train_loss'], label='PyTorch')
ax1.set_title('Training Loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.legend()

# Plot test accuracy
ax2.plot(mlx_metrics['test_accuracy'], label='MLX')
ax2.plot(torch_metrics['test_accuracy'], label='PyTorch')
ax2.set_title('Test Accuracy')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy')
ax2.legend()

# Plot epoch times
ax3.plot(mlx_metrics['epoch_times'], label='MLX')
ax3.plot(torch_metrics['epoch_times'], label='PyTorch')
ax3.set_title('Epoch Times')
ax3.set_xlabel('Epoch')
ax3.set_ylabel('Time (s)')
ax3.legend()

plt.tight_layout()
plt.show()
```

On the third image to the right, we can observe how MLX consistently takes less time on each epoch than PyTorch.
